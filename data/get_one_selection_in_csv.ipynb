```python
import requests
import pandas as pd
from pyjstat import pyjstat
import os

# URL of the JSON-stat data
url = "https://data.csu.gov.cz/api/dotaz/v1/data/vybery/CRU06AT1"

# Fetch the data
response = requests.get(url)
json_stat = response.json()

# Convert JSON-stat to pandas DataFrame
df = pyjstat.from_json_stat(json_stat)[0]

# Display the first few rows of the DataFro90ame
print("DataFrame shape:", df.shape)
df.head(1000)

# Generate output filename with timestamp
output_file = 'CRU06AT1.csv'

# Save DataFrame to CSV
df.to_csv(output_file, index=False, encoding='utf-8')
print(f"Data saved to: {output_file}")
```
